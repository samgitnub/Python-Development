```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from ipywidgets import widgets, interact
import os
```

```python
def generate_certificate(name, internship_title, date, output_path="certificate.pdf"):
    width, height = A4
    c = canvas.Canvas(output_path, pagesize=A4)

    # Add background or border if required (Optional)
    # c.rect(30, 30, width-60, height-60, stroke=1, fill=0)

    # Add Title
    c.setFont("Helvetica-Bold", 24)
    c.drawCentredString(width / 2.0, height - 100, "Certificate of Internship")

    # Add Name
    c.setFont("Helvetica", 18)
    c.drawCentredString(width / 2.0, height - 180, f"This is to certify that")

    c.setFont("Helvetica-Bold", 20)
    c.drawCentredString(width / 2.0, height - 220, name)

    # Add Internship Details
    c.setFont("Helvetica", 18)
    c.drawCentredString(width / 2.0, height - 260, f"has successfully completed an internship in")
    c.drawCentredString(width / 2.0, height - 300, f"{internship_title}")

    # Add Date
    c.setFont("Helvetica", 16)
    c.drawCentredString(width / 2.0, height - 360, f"Date: {date}")

    # Add Footer (Optional)
    c.setFont("Helvetica-Oblique", 12)
    c.drawCentredString(width / 2.0, 50, "Thank you for your contribution!")

    # Save PDF
    c.save()
    print(f"Certificate saved as: {output_path}")
```

```python
def generate_and_display(name, internship_title, date):
    output_path = f"{name}_certificate.pdf"
    generate_certificate(name, internship_title, date, output_path)
    return output_path

def certificate_ui():
    name_input = widgets.Text(description="Name:")
    title_input = widgets.Text(description="Title:")
    date_input = widgets.Text(description="Date:")
    output_message = widgets.Output()

    def on_submit(_):
        with output_message:
            output_message.clear_output()
            output_path = generate_and_display(name_input.value, title_input.value, date_input.value)
            print(f"Certificate generated: {output_path}")

    submit_button = widgets.Button(description="Generate Certificate")
    submit_button.on_click(on_submit)

    display(widgets.VBox([name_input, title_input, date_input, submit_button, output_message]))

certificate_ui()
```
